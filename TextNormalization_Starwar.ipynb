```python
# main lib
import glob
import string
import pickle 

# NLP lib
import nltk
from nltk.corpus import stopwords
nltk.download('wordnet') 
from nltk.stem import WordNetLemmatizer 
```

```python
# get Starwar file paths (ep4, ep5, ep6)
files = glob.glob('C:\\Users\\Supalerk\\Desktop\\Streaming_assignment\\*.txt')
files
```

```python
# remove stop word
stp_words = stopwords.words('english')

# text normalization using lemmatizer
wnl = WordNetLemmatizer() 

# create dictionary which key : charactor, value : message documents
final_dic = {}

# iterate through every file in files (ep4, ep5, ep6)
for file in files:
    
    # open file and readlines
    with open(file, 'r') as f:
        temp_txt = f.readlines()
        
        # ignore header
        temp_txt = temp_txt[1:]
        
    # iterate through every lines
    for txt in temp_txt:
        
        # split and assign type
        idx, char, message = txt.split(' ',2)
        # clean charactor name
        char = char.replace('"','')
        # remove punctuation from message
        message = message.translate(str.maketrans('', '', string.punctuation))
        message = message.replace('\n','')
        
        # remove empty str, stopwords and lower words
        clean_message = [ w.lower() for w in message.split(' ') if w not in stp_words and len(w)>=1]
        # text normalization with lemmatizer
        clean_message = [wnl.lemmatize(w) for w in clean_message]
        
        # if charactor occur before then append list of message to dictionary value
        # else create list of message for new charactor
        if char in final_dic:
            final_dic[char] += [clean_message]
        else:
            final_dic[char] = [clean_message]
        
```

```python
file_name = 'starwar_content.plk'

# dump final_dic to starwar_content.plk
with open(file_name, 'wb') as f:
    pickle.dump(final_dic, f, protocol=pickle.HIGHEST_PROTOCOL)
```
